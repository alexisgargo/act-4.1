fuente: https://www.kaggle.com/datasets/danagerous/sleep-data/data

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

with open('sleepdata.csv', 'r') as file:
    reader = csv.reader(file)
    column_data = []
    for row in reader:
        try:
            # los valores del csv estan en la siguiente forma:
            # ['2017-07-11 21:39:56;2017-07-12 06:06:25;65%;8:26;;;;5504']
            # horas de sueno estan en la cuarta columna, es necesario
            # transformar el tiempo de XX:XX a float
            value = (float(row[0].split(';')[3].split(':')[0]) +
                     float(row[0].split(';')[3].split(':')[1]) / 60)

        except (IndexError, ValueError):
            # algunas columnas estan vacias
            continue

        column_data.append(value)

data = np.array(column_data)
print('# de datos: ' + str(len(data)))

# maximos y minimos
print('min: ' + str(np.min(data)))
print('max: ' + str(np.max(data)))

# media y desviacion estandar
print('media: ' + str(np.mean(data)))
print('desviacion estandar: ' + str(np.std(data)))

plt.hist(data, bins=20)
plt.show()
```

El histograma de las horas de sueno tiene una forma que se asemeja a una distribucion normal, con un pico principal alrededor de las 8 horas, especificamente la media es de 7.63 horas.

La distribución es ligeramente asimetrica con algunos valores registrados en el extremo izquierdo

```python
from fitter import Fitter
f = Fitter(data)
f.fit()
f.summary()
```

fitter muestra la distribucion t, generalmente usada en muestras pequenas, como la mas cercana por tener el menor error cuadratico, posiblemente debido a los valores del extremo izquierdo.

Los parametros a considerar en una distribucion t incluyen los grados de libertad, que determinan el peso de las colas, y la localizacion donde se centra la distribucion

Volviendo a la distribucion normal sus parametros fundamentales son la media y la desviacion estandar. En este caso, la media seria aproximadamente 7.63 horas y la desviacion estandar 1.31.

Algunos usos de un modelo basado en la distribucion normal para horas de sueño podrian ser establecer umbrales estadisticamente fundamentados para definir el sueño "normal", "insuficiente" o "excesivo", comparar el patron de sueño de un individuo con una poblacion o evaluar la efectividad de ciertos cambios para modificar patrones de sueño.

La distribucion normal es particularmente util para identificar similitudes entre procesos. La prevalencia de la distribucion normal en fenomenos naturales  permite encontrar conexiones entre procesos que comparten esta caracteristica. Procesos biologicos como el sueño, la presion arterial, la altura o el peso tienden a seguir aproximadamente distribuciones normales, sugiriendo principios regulatorios comunes. Comparar las medias y desviaciones estandar entre diferentes poblaciones o procesos permite cuantificar similitudes y diferencias de manera estandarizada.
